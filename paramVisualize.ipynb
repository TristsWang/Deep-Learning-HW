```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from MyNN import load_model 

model = load_model('model.pkl')

# load the model
W1, W2, W3 = model.W1, model.W2, model.W3
a0, a1, a2, a3 = model.a

input_dim = 784  # 输入是28x28的图像，即784个像素点  
hidden1_dim = model.hidden1_dim 
hidden2_dim =  model.hidden2_dim  
output_dim = 10  # 有10个类别  
    
# 权重矩阵热力图可视化（接受轴对象作为参数）  
def plot_heatmap(ax, W, title):  
    cax = ax.matshow(W, cmap='viridis')  
    fig = ax.get_figure()  # 获取当前图形对象  
    fig.colorbar(cax, ax=ax, location='bottom')  # 在当前轴上添加颜色条，并设置其位置  
    ax.set_title(title)  

# 创建一个1行3列的子图网格  
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))  # 你可以根据需要调整figsize  
  
# 在每个子图上绘制热力图  
plot_heatmap(axes[0], W1, 'Weight Matrix W1 (Input~Hidden1)')  
plot_heatmap(axes[1], W2, 'Weight Matrix W2 (Hidden1~Hidden2)')  
plot_heatmap(axes[2], W3, 'Weight Matrix W3 (Hidden2~Output)')  
  
# 显示图形  
plt.tight_layout()  # 调整子图参数，使之填充整个图像区域  

plt.show()
```

```python
import numpy as np  
import matplotlib.pyplot as plt  
from sklearn.decomposition import PCA  
  
# 隐藏层输出降维可视化  
def plot_pca_reduction(ax, W, title):  
    pca = PCA(n_components=2)  # 降维到2维  
    X_pca = pca.fit_transform(W)  
    ax.scatter(X_pca[:, 0], X_pca[:, 1])  # 绘制降维后的数据点  
    ax.set_xlabel('PCA Feature 1')  # 设置x轴标签  
    ax.set_ylabel('PCA Feature 2')  # 设置y轴标签  
    ax.set_title(title)  # 设置标题  
  
# 创建一个1行3列的子图网格  
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))  # 你可以根据需要调整figsize  
  
# 绘制隐藏层输出的降维结果  
plot_pca_reduction(axes[0,0], a0, 'PCA Reduction of input')  
plot_pca_reduction(axes[0,1], a1, 'PCA Reduction of hidden_layer1 output')  
plot_pca_reduction(axes[1,0], a2, 'PCA Reduction of hidden_layer2 output')  
plot_pca_reduction(axes[1,1], a3, 'PCA Reduction of output')  
  
# 显示图形  
plt.tight_layout()  # 调整子图参数，使之填充整个图像区域且不会重叠  
plt.show()
```
